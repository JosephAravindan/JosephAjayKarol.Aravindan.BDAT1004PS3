# Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
import pandas as pd

us_marriage_divorce_data = pd.read_csv('C:\\Users\\josep\\Downloads\\us-marriages-divorces-1867-2014.csv')
years = us_marriage_divorce_data['Year'].values
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values

df_marraige_divorce = pd.DataFrame(list(zip( marriages_per_capita , divorces_per_capita )),
                  index = years , 
                  columns = [ 'marriages_per_capita' , 'divorces_per_capita' ])

df_marraige_divorce
```

```python
df_marraige_divorce.plot(xlabel='YEAR', ylabel='Count Per Capita',figsize=(20,10), title='Marriage and Divorce per Capita in US - 1867 to 2014')
```
