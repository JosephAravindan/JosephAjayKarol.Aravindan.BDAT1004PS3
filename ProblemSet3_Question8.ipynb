# Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

hollywood_actor_killed = pd.read_csv('C:\\Users\\josep\\Downloads\\actor_kill_counts.csv')
sortData=hollywood_actor_killed.sort_values('Count', ascending = True)

actor_names = sortData['Actor'].values
kill_counts = sortData['Count'].values
c=["red"]
plt.barh(actor_names,kill_counts,color=c)
plt.xlabel("Number of People Killed")
plt.ylabel("Actor")
plt.title("Deadliest Actors in Hollywood")

plt.show()
```
